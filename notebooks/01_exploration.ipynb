```python
from pathlib import Path

data_dir = Path("..") / "data" / "raw" / "EuroSAT" / "2750"
data_dir.exists(), data_dir
```

```python
class_dirs = [p for p in data_dir.iterdir() if p.is_dir()]
[len(class_dirs), [p.name for p in class_dirs]]
```

```python
for class_dir in class_dirs:
    images = list(class_dir.glob("*"))
    print(class_dir.name, len(images))
```

```python
# Show sample images from one class

import matplotlib.pyplot as plt
from PIL import Image

sample_class = class_dirs[0]
sample_images = list(sample_class.glob("*"))[:4]

fig, axes = plt.subplots(1, 4, figsize=(12, 3))
for ax, img_path in zip(axes, sample_images):
    img = Image.open(img_path)
    ax.imshow(img)
    ax.set_title(sample_class.name)
    ax.axis("off")

plt.tight_layout()
```
